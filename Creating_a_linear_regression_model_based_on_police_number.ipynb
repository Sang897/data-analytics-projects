```python
letspredict
import pandas as pd 
import numpy as np 
from sklearn import linear_model
import matplotlib.pyplot as plt
import io
```

```python
policeprediction = pd.read_csv(io.BytesIO(letspredict['Predictcrimenumber.csv'])) #This converts the data to be readable dataframe. 
```

```python
# Create a linear regression line: 
regressionmodel = linear_model.LinearRegression()
regressionmodel.fit(new_df,Crimenumber)
```
