```python
#The objective of this project is to see which types of app( )was used mostly to comment about the netflix drama Squidgame. 


import pandas as pd
import os
print(os.getcwd())
```

```python
os.chdir('C:\\Users\\spark\\Desktop\\Data Science Projects\\Squid Game Analysis') #This is changing the working directory
squidgame = pd.read_csv('Datascienceproject_1.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns #Uses matplotlib but works as an upgrade of matplotlib. 
%matplotlib inline 
```

```python
squidgame=squidgame[0:99]
```

```python
sns.catplot( x='source',
             kind="count", 
             height=15, 
             aspect=3, 
             data=squidgame)
plt.xlabel("Education", size=20)
plt.ylabel("Count", size=20)
plt.title("Cool")
plt.tight_layout()
plt.savefig("Countplot_or_barplot_with_Seaborn_catplot.png")
```

```python
plt.figure(figsize=(50,1))
```

```python
squidgame.head()
```

```python
squidgame.drop(columns=["source"],axis=1,inplace=False)
```

```python
squidgame.head()
```

```python
indexNames = squidgame[squidgame['source'] == 'FALSE'].index
indexNames2 = squidgame[squidgame['source'] == 'PromoRepublic'].index
indexNames3 = squidgame[squidgame['source'] == 'Buffer'].index
indexNames4 = squidgame[squidgame['source'] == 'Twitter for Ipad'].index
indexNames5 = squidgame[squidgame['source'] == 'TweetDeck'].index
```

```python
squidgame.drop(indexNames, inplace=True)
squidgame.drop(indexNames2, inplace=True)
squidgame.drop(indexNames3, inplace=True)
squidgame.drop(indexNames4, inplace=True)
squidgame.drop(indexNames5, inplace=True)
```

```python
sns.catplot( x='source',
             kind="count", 
             height=15, 
             aspect=5, 
             data=squidgame)
plt.xlabel("Education", size=20)
plt.ylabel("Count", size=20)
plt.title("Cool")
plt.tight_layout()
plt.savefig("Countplot_or_barplot_with_Seaborn_catplot.png")
```
